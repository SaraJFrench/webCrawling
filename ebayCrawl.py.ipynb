```python
import requests
from bs4 import BeautifulSoup
import csv


def get_page(url):
    response = requests.get(url)

    if not response.ok:
        print('stat:', response.status_code)
    else:
        soup = BeautifulSoup(response.text, 'lxml')
    return soup


def get_detail_data(soup):
    # we want title, price, and items sold
    try:
        title = soup.find('h1', id='itemTitle').text.replace('\\xa0', '')
    except:
        title = ''
    # print(title)

    try:
        try:
            p = soup.find('span', id='prcIsum').text.strip()

        except:
            p = soup.find('span', id='prcIsum_bidPrice').text.strip()
        currency, price = p.split(' ')

    except:
        currency = ''
        price = ''
    # print(currency, price)

    try:
        sold = soup.find('span', class_='w2b-sgl').text.strip().split(' ')[0]
    except:
        sold = ''
    # print(sold)

    data = {
        'title': title,
        'price': price,
        'currency': currency,
        'total sold': sold
    }
    return data


def get_index_data(soup):
    try:
        links = soup.find_all('a', class_='s-item__link')
    except:
        links = []
    # print(links)
    urls = [item.get('href') for item in links]
    # print(urls)
    return urls


def write_csv(data, url):
    with open('output2.csv', 'a') as csvfile:
        writer = csv.writer(csvfile)

        row = [data['title'], data['price'], data['currency'], data['total sold'], url]

        writer.writerow(row)


def main():
       url = 'https://www.ebay.com/sch/i.html?_nkw=amazon+echo+dot&_png=1'
       #url = 'https://www.ebay.com/sch/i.html?_nkw=amazon+echo+dot'
       #url = 'https://www.ebay.com/itm/Amazon-Echo-Dot-3rd-Generation-w-Alexa-Voice-Media-Device-BRAND-NEW/283840411198?epid=26024048451&hash=item421634f63e%3Ag%3AAswAAOSwDQdejNOO&LH_BIN=1&LH_ItemCondition=1000'
       products = get_index_data(get_page(url))
       
      
       for link in products:
           data = get_detail_data(get_page(link))
           #print(data)
           write_csv(data, link)
        
       
if __name__ == '__main__':
    main()







```
